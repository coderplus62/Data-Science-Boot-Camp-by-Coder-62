# **Pengenalan tentang Matplotlib - Konsep Dasar**

Matplotlib adalah pustaka yang populer untuk visualisasi data di Python. Dengan Matplotlib, Anda dapat membuat berbagai jenis grafik dan plot untuk mengilustrasikan data secara visual. Mari kita mulai dengan memahami beberapa konsep dasar dalam Matplotlib.

## **Langkah 1.1: Axes (Sumbu)**
Dalam Matplotlib, konsep yang penting adalah "axes" (sumbu). Sumbu adalah area di dalam plot yang berisi data yang akan divisualisasikan. Anda dapat membuat sumbu dengan `plt.axes()`:

```python
import matplotlib.pyplot as plt

# Membuat sumbu dengan koordinat dari 0 hingga 1 di kedua sumbu
ax = plt.axes()
ax.plot([0, 1], [0, 1])
plt.show()
```

## **Langkah 1.2: Metode Axes vs. pyplot**

Dalam Matplotlib, Anda memiliki dua pendekatan untuk membuat plot: menggunakan metode "Axes" atau menggunakan "pyplot". Pendekatan menggunakan metode "Axes" memberi Anda lebih banyak kontrol atas tampilan plot:

```python
# Menggunakan metode Axes
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot([0, 1], [0, 1])
plt.show()
```

```python
# Menggunakan pyplot
plt.plot([0, 1], [0, 1])
plt.show()

# Pendekatan menggunakan "pyplot" lebih ringkas tetapi mungkin kurang fleksibel
```

## **Langkah 1.3: Beberapa Sumbu**

Anda dapat membuat beberapa sumbu dalam satu gambar untuk membuat plot yang lebih kompleks:

```python
# Membuat beberapa sumbu dalam satu gambar
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
axes[0, 0].plot([0, 1], [0, 1])
axes[0, 1].plot([1, 0], [0, 1])
axes[1, 0].plot([0, 1], [1, 0])
axes[1, 1].plot([1, 0], [1, 0])
plt.show()
```
