# **Kustomisasi Plot dengan Matplotlib**

Selain membuat plot dasar, Matplotlib juga memungkinkan Anda untuk menyesuaikan tampilan plot agar sesuai dengan preferensi Anda. Mari kita lihat cara melakukan kustomisasi pada plot menggunakan Matplotlib.

## **3.1 Warna, Gaya, dan Marker**

Anda dapat menyesuaikan warna, gaya garis, dan marker pada plot:

```python
import matplotlib.pyplot as plt

# Data untuk plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Kustomisasi plot
plt.plot(x, y, color='blue', linestyle='dashed', linewidth=2, marker='o', markersize=8, label='Garis A')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Kustomisasi Plot')
plt.legend()
plt.show()
```

## **3.2 Warna Latar Belakang dan Batas Plot**

Anda dapat mengatur warna latar belakang dan batas plot:

```python
# Data untuk plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Kustomisasi latar belakang dan batas plot
plt.plot(x, y, label='Garis A')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Kustomisasi Latar Belakang dan Batas Plot')
plt.legend()
plt.gca().set_facecolor('lightgray')  # Atur latar belakang
plt.gca().spines['top'].set_visible(False)  # Sembunyikan batas atas
plt.gca().spines['right'].set_visible(False)  # Sembunyikan batas kanan
plt.show()
```

## **3.3 Menambahkan Teks dan Anotasi**

Anda dapat menambahkan teks dan anotasi pada plot:

```python
# Data untuk plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Menambahkan teks dan anotasi
plt.plot(x, y, label='Garis A')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Menambahkan Teks dan Anotasi')
plt.legend()
plt.annotate('Titik Penting', xy=(3, 6), xytext=(4, 4), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.text(2, 9, 'Teks Tambahan', style='italic')
plt.show()
```
