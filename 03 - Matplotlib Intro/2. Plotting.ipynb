# **Visualisasi Data dengan Matplotlib - Plotting**

Matplotlib adalah alat yang kuat untuk membuat berbagai jenis plot dan visualisasi data. Mari kita lanjutkan dengan memahami cara membuat plot garis, scatter plot, dan histogram menggunakan Matplotlib.

## **Langkah 2.1: Line Plot (Plot Garis)**
Untuk membuat plot garis, Anda dapat menggunakan fungsi `plot()`:

```python
import matplotlib.pyplot as plt

# Data untuk plot garis
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat plot garis
plt.plot(x, y, label='Garis A')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Plot Garis')
plt.legend()
plt.show()
```

## **Langkah 2.2: Scatter Plot (Plot Sebar)**

Untuk membuat scatter plot, Anda juga dapat menggunakan fungsi `scatter()`:

```python
# Data untuk scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat scatter plot
plt.scatter(x, y, label='Data Sebar', color='red', marker='o')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

## **Langkah 2.3: Histogram Plot**

Anda dapat membuat histogram menggunakan fungsi `hist()`:

```python

# Data untuk histogram
data = [10, 15, 20, 25, 30, 30, 35, 40, 45, 50, 55, 60, 65]

# Membuat histogram
plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram Plot')
plt.show()
```
