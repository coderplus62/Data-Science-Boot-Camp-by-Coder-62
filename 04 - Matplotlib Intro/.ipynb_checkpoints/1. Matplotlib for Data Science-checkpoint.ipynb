# 1. Fundamental Concept

Matplotlib adalah pustaka yang populer untuk visualisasi data di Python. Dengan Matplotlib, Anda dapat membuat berbagai jenis grafik dan plot untuk mengilustrasikan data secara visual. Mari kita mulai dengan memahami beberapa konsep dasar dalam Matplotlib.

## 1.1. Axes (Sumbu)
Dalam Matplotlib, konsep yang penting adalah "axes" (sumbu). Sumbu adalah area di dalam plot yang berisi data yang akan divisualisasikan. Anda dapat membuat sumbu dengan `plt.axes()`:

```python
import matplotlib.pyplot as plt

# Membuat sumbu dengan koordinat dari 0 hingga 1 di kedua sumbu
ax = plt.axes()
ax.plot([0, 1], [0, 1])
plt.show()
```

## 1.2. Metode Axes vs. pyplot

Dalam Matplotlib, Anda memiliki dua pendekatan untuk membuat plot: menggunakan metode "Axes" atau menggunakan "pyplot". Pendekatan menggunakan metode "Axes" memberi Anda lebih banyak kontrol atas tampilan plot:

```python
# Menggunakan metode Axes
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot([0, 1], [0, 1])
plt.show()
```

```python
# Menggunakan pyplot
plt.plot([0, 1], [0, 1])
plt.show()

# Pendekatan menggunakan "pyplot" lebih ringkas tetapi mungkin kurang fleksibel
```

## 1.3. Beberapa Sumbu

Anda dapat membuat beberapa sumbu dalam satu gambar untuk membuat plot yang lebih kompleks:

```python
# Membuat beberapa sumbu dalam satu gambar
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
axes[0, 0].plot([0, 1], [0, 1])
axes[0, 1].plot([1, 0], [0, 1])
axes[1, 0].plot([0, 1], [1, 0])
axes[1, 1].plot([1, 0], [1, 0])
plt.show()
```

# 2. Plotting - Visualisasi Data dengan Matplotlib

Matplotlib adalah alat yang kuat untuk membuat berbagai jenis plot dan visualisasi data. Mari kita lanjutkan dengan memahami cara membuat plot garis, scatter plot, dan histogram menggunakan Matplotlib.

## 2.1. Line Plot (Plot Garis)
Untuk membuat plot garis, Anda dapat menggunakan fungsi `plot()`:

```python
import matplotlib.pyplot as plt

# Data untuk plot garis
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat plot garis
plt.plot(x, y, label='Garis A')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Plot Garis')
plt.legend()
plt.show()
```

## 2.2. Scatter Plot (Plot Sebar)

Untuk membuat scatter plot, Anda juga dapat menggunakan fungsi `scatter()`:

```python
# Data untuk scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat scatter plot
plt.scatter(x, y, label='Data Sebar', color='red', marker='o')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

## 2.3. Histogram Plot

Anda dapat membuat histogram menggunakan fungsi `hist()`:

```python

# Data untuk histogram
data = [10, 15, 20, 25, 30, 30, 35, 40, 45, 50, 55, 60, 65]

# Membuat histogram
plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.title('Histogram Plot')
plt.show()
```

# 3. Kustomisasi Plot dengan Matplotlib

Selain membuat plot dasar, Matplotlib juga memungkinkan Anda untuk menyesuaikan tampilan plot agar sesuai dengan preferensi Anda. Mari kita lihat cara melakukan kustomisasi pada plot menggunakan Matplotlib.

## 3.1. Color, Style, dan Marker

Anda dapat menyesuaikan warna, gaya garis, dan marker pada plot:

```python
import matplotlib.pyplot as plt

# Data untuk plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Kustomisasi plot
plt.plot(x, y, color='blue', linestyle='dashed', linewidth=2, marker='o', markersize=8, label='Garis A')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Kustomisasi Plot')
plt.legend()
plt.show()
```

## 3.2. Warna Latar Belakang dan Batas Plot

Anda dapat mengatur warna latar belakang dan batas plot:

```python
# Data untuk plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Kustomisasi latar belakang dan batas plot
plt.plot(x, y, label='Garis A')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Kustomisasi Latar Belakang dan Batas Plot')
plt.legend()
plt.gca().set_facecolor('lightgray')  # Atur latar belakang
plt.gca().spines['top'].set_visible(False)  # Sembunyikan batas atas
plt.gca().spines['right'].set_visible(False)  # Sembunyikan batas kanan
plt.show()
```

## 3.3. Menambahkan Teks dan Anotasi

Anda dapat menambahkan teks dan anotasi pada plot:

```python
# Data untuk plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Menambahkan teks dan anotasi
plt.plot(x, y, label='Garis A')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Menambahkan Teks dan Anotasi')
plt.legend()
plt.annotate('Titik Penting', xy=(3, 6), xytext=(4, 4), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.text(2, 9, 'Teks Tambahan', style='italic')
plt.show()
```
